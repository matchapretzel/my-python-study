```python
def is_divisible_by_7_not_5():
    num=int(input('enter a number:'))
    if num%7 == 0 and num%5 != 0:
        result=True
    else:
        result=False
    return result
is_divisible_by_7_not_5()
```

逻辑运算符 and / or 不仅返回 True 或 False，它们返回的是能确定结果的那个操作数的实际值。

规则：

a or b：如果 a 为真，返回 a 的值， __（短路）__ 否则返回 b 的值。

a and b：如果 a 为假，返回 a 的值， __（短路）__ 否则返回 b 的值。

```python
#Example of which above
False or True and print('here it comes!')

#False or ... 不会短路，所以要计算右边 (True and print("Hello"))
#True and ... 不会短路，所以要计算 print("Hello")
#print 执行，返回 None
#True and None → None
#False or None → None
#所以 result 为 None，但 print 执行了。
```

在编程中，一个函数执行时可以做两件事：  
1.执行一些操作（比如在屏幕上显示文字、修改文件等）  
2.__产生一个结果值，这个结果值就是函数的“返回值”（return value）。__  
当你调用一个函数时，它通常会返回一个值给调用处。  
例如：  
python  
x = len("abc")  # len 函数返回 3  
这里 len 函数返回了 3，然后赋值给 x。

```python
a=float(input('enter the first number:'))
b=str(input('how to calculate?'))
c=float(input('enter the second number:'))


if b == '+':
    result=a+c
elif b == '-':
    result=a-c
elif b == '*':
    result=a*c
elif b == '/' and c!=0:
    result=a/c
else:
    result='Error'

print(result)
```
