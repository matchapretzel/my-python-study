```python
#水仙数 通用位数解法 创造出来去匹配
def find_narcissistic_numbers(max_digits=7):
    """
    查找指定位数范围内的水仙花数
    
    Args:
        max_digits: 最大位数（包含）
    
    Returns:
        所有找到的水仙花数列表
    """
    narcissistic_numbers = []
    
    for digits in range(1, max_digits + 1):
        # 计算当前位数下的最小和最大值
        start = 10 ** (digits - 1)
        end = 10 ** digits
        
        # 特殊情况：1位数时，start应该是1
        if digits == 1:
            start = 0
        
        # 计算每个数字的n次幂，提高效率
        powers = [i ** digits for i in range(10)]
        
        for num in range(start, end):
            temp = num
            digit_sum = 0
            
            # 分解数字并计算幂和
            while temp > 0:
                digit_sum += powers[temp % 10]
                temp //= 10
            
            if digit_sum == num:
                narcissistic_numbers.append(num)
    
    return narcissistic_numbers
```
